```python
from dataclasses import dataclass

joint_list = []
bvh = []
Frames = []
Frame_Num = 0
Frame_Time = 0


@dataclass
class Node:
    name : str
    offset : list
    channels : dict
    parent : str
    child : list
    ID : int


with open('data/try.bvh', 'r') as f:
    for line in f.readlines():
        a = line.split()
        bvh.append(a)

# 入栈一次 id++, 出栈一次id--,可以给上一个节点赋予孩子
id = 0
now_line = 0
for line in range(len(bvh)):
    now_line = line
    this_line = bvh[line]
    print(this_line)
    if(this_line[0] == "HIERARCHY"):
        continue
    if(this_line[0] == "ROOT"):
        joint_list.append(Node(
            name = this_line[1],
            offset= [],
            channels= {},
            parent = None,
            child = [],
            ID = id
        ))
        continue
    if(this_line[0] == '{'):
        id += 1 # next node id
        continue
    if(this_line[0] == "OFFSET"):
        joint_list[id - 1].offset = [x for x in this_line[1:]]
        continue
    if(this_line[0] == 'CHANNELS'):
        joint_list[id - 1].channels = {
            "channel_nums" : this_line[1],
            "channels" : this_line[2:]
        }
        continue
    if(this_line[0] == "JOINT"):
        joint_list.append(Node(
            name = this_line[1],
            offset= [],
            channels= [],
            parent = joint_list[id - 1].name,
            child = [],
            ID = id
        ))
        joint_list[id - 1].child.append(joint_list[id].name)
        continue
    if(this_line[0] == '}'):
        id -= 1
        continue
    if(this_line[0] == 'End'):
        print("---")
        print(bvh[now_line + 1])
        print(bvh[now_line + 2])
        print(bvh[now_line + 3])
        print("---")
        print(id)
        joint_list.append(Node(
            name = this_line[1] + "_end_" + str(id),
            offset = bvh[now_line + 2][1:],
            channels = None,
            parent = joint_list[id - 1].name,
            child = None,
            ID = id
        ))
        id -= 1
        line += 3
        continue
    if(this_line[0] == 'MOTION'):
        continue
    if(this_line[0] == 'Frames:'):
        Frame_Num = int(this_line[1])
        continue
    if(this_line[0] == "Frame"):
        Frame_Time = float(this_line[2])
        continue
    Frames.append(this_line)
    

```
