```python
import os

def generate_tree(dir_path, prefix=""):
    # 获取指定目录中的所有文件和子目录列表
    contents = os.listdir(dir_path)
    
    # 指定指针列表，用于生成目录树的符号
    pointers = [ '├── ' ] * (len(contents) - 1) + [ '└── ' ]
    #print(pointers)
    
    # 迭代目录中的每个文件和子目录
    for pointer, content in zip(pointers, contents):
        # 构造当前文件或子目录的完整路径
        path = os.path.join(dir_path, content)
        
        # 打印当前文件或子目录的名称，带有适当的前缀和指针符号
        print(prefix + pointer + content)
        
        # 如果当前路径是一个目录，则递归调用 generate_tree 函数
        if os.path.isdir(path):
            # 根据指针符号设置扩展前缀
            extension = '│   ' if pointer == '├── ' else '    '
            
            # 递归调用 generate_tree 函数以处理子目录
            generate_tree(path, prefix + extension)

# 使用示例
generate_tree('D:\BaiduSyncdisk\games105\GAMES105-作业\GAMES105-作业\lab1')
```
